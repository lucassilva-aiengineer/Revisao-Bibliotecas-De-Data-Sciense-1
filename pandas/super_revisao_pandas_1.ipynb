```python
from datetime import datetime, timedelta
```

```python
datetime.now() - timedelta(10)
```

```python
def gerar_datetime(data):
  hoje = datetime.now()

  return hoje - timedelta(data)
```

```python
# Gerando um Datetime

df["data_time"] = [gerar_datetime(dias) for dias in range(1000, 0, -1)]
```

```python
df.head()
```

```python
df.info()
```

```python
# Passando um data string para o formato DataFrame

df["data_timestamp"] = pd.to_datetime(df["data_timestamp"], format= '%Y-%m-%d')
```
