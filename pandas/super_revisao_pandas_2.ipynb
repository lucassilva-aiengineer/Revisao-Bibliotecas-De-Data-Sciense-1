```python
import pandas as pd
import random
```

```python
# Séries

# Séries são uma estrutura unidimensional que se assemelha muito aos
# vetores (listas python).

valores = ["a", "b", "c", "d", "e", "f"]

serie_a = pd.Series(valores)
```

```python
serie_a
```

```python
def criar_series_1()-> pd.Series:

  vetores = [random.randint(0, 1000) for a  in range(100)]
  serie = pd.Series(vetores)

  return serie
```

```python
minha_serie = criar_series_1()
```

```python
minha_serie.iloc[0]
```

```python
# Criando séries com índice

vetor = ["Marcos", "João", "Carlos", "Mateus", "Lucas"]

serie_2 = pd.Series(vetor, index=["nome1", "nome2", "nome3", "nome4", "nome5"])
serie_2
```

```python
# Acessando valores pelo índice
print(serie_2.loc["nome1"])
```

```python
# DataFrames

# DataFrames são estrutura de dados bidimensionais
# Temos linhas e colunas, colunas destinadas a valores específicos
# E a cada uma destas linhas um valor

dicionario = {
    "nomes": ["Marcos", "Ronaldo", "José", "Jonathas", "Matheus"],
    "idades": [20, 22, 23, 19, 18],
    "cargo": ["cargo1", "cargo1", "cargo2", "cargo1", "cargo2"]
}

df_1 = pd.DataFrame(dicionario)
```

```python
# Definindo uma coluna específica no lugar da coluna de ínidice

df_1 = df_1.set_index("cargo")
```

```python
df_1.loc["cargo1"]
```

```python
# Somando dois DataFrames

df_a = pd.DataFrame({"coluna_a": [10, 30, 55]})

df_b = pd.DataFrame({"coluna_a": [20, 35, 42]})

df_a + df_b
```

```python
# Reiniciano o ínidice

df_1 = df_1.reset_index()
```

```python
df_3 = df_a + df_b
```

```python
df_3
```

```python
# Salvando um DataFrame

df_1.to_csv("meu_dataframe.csv", index= False)
```

```python
# Transformando um DataFrame em um dicionário.
df_1.to_dict()
```
